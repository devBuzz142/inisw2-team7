```python
from moviepy.editor import VideoFileClip, TextClip, CompositeVideoClip
from moviepy.video.tools.subtitles import SubtitlesClip
```

```python
# 자막 파일 경로
srt_path = "C:/Users/ella6/Downloads/srtSubtitles.srt"
# 동영상 파일 경로
video_path = "C:/Users/ella6/Videos/video_out.avi"
# 동영상 로드
video = VideoFileClip(video_path)
# 동영상 추출
#output_path="C:/Users/ella6/Videos/output9.mp4"
```

```python
import pickle

with open("faces.pckl", 'rb') as f:
    faces = pickle.load(f)
    for face in faces:
        print(face)
```

```python
def time_to_seconds(time_str):
    h, m, s = map(int, time_str.split(':'))
    seconds = h * 3600 + m * 60 + s
    return seconds
```

```python
import re

def extract_time(timestamp):
    pattern = r'(\d{2}:\d{2}:\d{2}.\d{3})\s-->\s(\d{2}:\d{2}:\d{2}.\d{3})'
    matches = re.findall(pattern, timestamp)
    if matches:
        start_time, end_time = matches[0]
        return start_time, end_time
    else:
        return None, None

#subtitle_path = "subtitle.txt"  # 자막 파일 경로

subtitles_clip = []

with open(srt_path, "r") as f:
    lines = f.readlines()
    
for i in range(0, len(lines), 4):
    start_time, end_time = lines[i + 1].strip().split(' --> ')
    text = lines[i + 2].strip()
    if text == '':
        continue
    text_clip = TextClip(text, fontsize=24, color='white', font='Arial').set_position(("center", "bottom")).set_start(start_time).set_end(end_time)
    start = int(time_to_seconds(start_time.split(',')[0])*video.fps)
    end = int(time_to_seconds(end_time.split(',')[0])*video.fps)
    mid = int((start+end)//2)
    if [d['bbox'] for d in faces[mid]] == []:
        text_clip = text_clip.set_position(("center", "bottom"))
        text_clip = text_clip.set_start(start_time).set_end(end_time)
    else:
        x, y = [d['bbox'] for d in faces[mid]][0][0], [d['bbox'] for d in faces[mid]][0][3]
        y += 10
        text_clip = text_clip.set_position([x, y])
        text_clip = text_clip.set_start(start_time).set_end(end_time)
    subtitles_clip.append(text_clip)
```

```python
# 텍스트 클립들을 동영상 위에 합성
video_with_text = CompositeVideoClip([video] + subtitles_clip)
```

```python
# 합성된 동영상을 저장
output_path="C:/Users/ella6/Videos/output14.mp4"
video_with_text.write_videofile(output_path, codec='libx264', audio_codec='aac')
```
